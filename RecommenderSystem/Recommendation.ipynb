```python
import ast

def getValue(obj):
    values=[]
    for i in ast.literal_eval(obj):
        values.append(i['name'])
    return values
```

```python
movies['genres'].apply(getValue)
```

```python
movies['genres']=movies['genres'].apply(getValue)
```

```python
movies['keywords']=movies['keywords'].apply(getValue)
```

```python
import ast

def getCast(obj):
    values=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter <3:
            values.append(i['name'])
        counter+=1
    return values
```

```python
movies['cast']=movies['cast'].apply(getCast)
```

```python
def getDirector(obj):
    director= []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            director.append(i['name'])
    return director 
```

```python
movies['crew']=movies['crew'].apply(getDirector)
```

```python
def Replace(text):
    values=[]
    for i in text:
        values.append(i.replace(" ",""))
    return values
```

```python
movies['cast']=movies['cast'].apply(Replace)
movies['crew']=movies['crew'].apply(Replace)
movies['genres']=movies['genres'].apply(Replace)
movies['keywords']=movies['keywords'].apply(Replace)
```

```python
movies['overview'].dtype
```

```python
movies['overview']=movies['overview'].apply(lambda x:str(x).split())
```

```python
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
movies.drop(['keywords','overview','cast','crew','genres'],axis=1,inplace=True)
```

```python
movies['tags']=movies['tags'].apply(lambda x:" ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
movies.iloc[6]
```

Recommender System using Cosine Similarity

```python
def Recommend(title):
    index=movies[movies['title']==title].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
           print("\t",movies.iloc[i[0]].title)
    
```

```python
print("Movies You May Like ...")
Recommend('Batman')
```

```python
import pickle
pickle.dump(similarity,open('Recommend.pkl','wb'))
```

Recommender System using KNN algorithm

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors()
model.fit(vectors)

input=vectors[1360]  #Batman
similar_movie=model.kneighbors([input],return_distance=False)[0]
movies.iloc[similar_movie]
```

```python
import pickle
pickle.dump(model,open('KnnSystem.pkl','wb'))
```
